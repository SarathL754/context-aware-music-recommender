```python
import re

def normalize(text):
    return re.sub(r'\s+', ' ', text.strip().lower())
```

```python
import random
import gymnasium as gym
import networkx as nx
from gymnasium import spaces

class SongRecEnv(gym.Env):
    def __init__(self, graph, users):
        self.G = graph
        self.users = users
        self.node_list = list(graph.nodes)
        self.node_index = {n: i for i, n in enumerate(self.node_list)}
        self.observation_space = gym.spaces.Discrete(len(self.node_list))
        self.action_space = gym.spaces.Discrete(len(self.node_list))
        self.current_user = None
        self.state = None
        self.hop = 0

    def reset(self, *, seed=None, user=None):
        if user is None:
            self.current_user = random.choice(self.users)
        else:
            self.current_user = user

        self.state = self.node_index[self.current_user]
        self.hop = 0

        return self.state, {}  # <- now returns a tuple (obs, info)


    def step(self, action):
        done = False
        reward = 0

        current_node = self.node_list[self.state]
        action_node = self.node_list[action]

        # Check if action is a valid successor
        if action_node not in self.G.successors(current_node):
            reward = -1
            done = True
            return self.state, reward, done, False, {}

        next_type = self.G.nodes[action_node]['type']

        if self.hop == 0 and next_type == "mood":
            self.state = self.node_index[action_node]
            self.hop += 1
            return self.state, reward, done, False, {}

        elif self.hop == 1 and next_type == "song":
            # Compare moods

            user_mood = normalize(list(self.G.successors(self.current_user))[0])
            song_mood = normalize(list(self.G.predecessors(action_node))[0])
            reward = 1 if user_mood == song_mood else 0
            done = True
            self.state = self.node_index[action_node]
            return self.state, reward, done, False, {}

        else:
            reward = -1
            done = True
            return self.state, reward, done, False, {}

    def render(self, mode="human"):
        print(f"Current state: {self.node_list[self.state]}")
    def close(self):
        pass
```

```python
# Assuming G is your DiGraph, and users is your user node list
env = SongRecEnv(graph=G, users=[n for n, d in G.nodes(data=True) if d['type'] == 'user'])


from stable_baselines3.common.env_checker import check_env
check_env(env)

from stable_baselines3 import DQN

model = DQN("MlpPolicy", env, verbose=1, learning_rate=0.001, buffer_size=10000, exploration_fraction=0.1)
model.learn(total_timesteps=100000)
```

```python
obs, info = env.reset()
terminated = False
truncated = False

while not (terminated or truncated):
    action, _ = model.predict(obs)
    obs, reward, terminated, truncated, info = env.step(action)
    env.render()

print(f"Reward: {reward}")

```

```python
import random
import networkx as nx

def format_path_with_edges(G, path):
    result = []
    for i in range(len(path)-1):
        u, v = path[i], path[i+1]
        edge_data = G.get_edge_data(u, v)
        relation = edge_data.get("relation", "?") if edge_data else "?"
        result.append(f"{u} --[{relation}]--> {v}")
    return "\n".join(result)
```

```python
paths = []

for user in df_users['user_id']:
    obs, info = env.reset(user=user)  # Now this works!
    path = [user]
    terminated = False
    truncated = False

    while not (terminated or truncated):
        action, _ = model.predict(obs)
        obs, reward, terminated, truncated, info = env.step(action)
        path.append(env.node_list[env.state])

    if reward == 1:  # Only successful mood-matched paths
        paths.append(path)
```

```python
ground_truth = {}

for _, row in df_users.iterrows():
    user_id = row['user_id']
    user_mood = row['music_Influencial_mood']

    # Songs whose predicted mood matches user mood
    gt_songs = list(df_songs[df_songs['predicted_mood_mistral'] == user_mood]['song_id'])

    ground_truth[user_id] = gt_songs
```

```python
def hit_at_k(recommended, ground_truth, k=5):
    return int(any(song in ground_truth for song in recommended[:k]))

# Evaluate Hit@1 for all users
hit_scores = []

for path in paths:
    user = path[0]
    recommended_songs = [node for node in path if node in df_songs['song_id'].values]

    if user in ground_truth:
        hit = hit_at_k(recommended_songs, ground_truth[user], k=1)
        hit_scores.append(hit)

# Final Hit@1 score
hit_at_1_score = sum(hit_scores) / len(hit_scores) if hit_scores else 0
print(f"🎯 Hit@1: {hit_at_1_score:.2f}")
```

```python
def recall_at_k(recommended, ground_truth, k=5):
    if not ground_truth:
        return 0
    hits = sum(1 for song in recommended[:k] if song in ground_truth)
    return hits / len(ground_truth)
```

```python
def precision_at_k(recommended, ground_truth, k=5):
    """Precision@K: proportion of recommended songs that are relevant (in ground truth)"""
    hits = [song for song in recommended[:k] if song in ground_truth]
    return len(hits) / k
```

```python
import numpy as np

def ndcg_at_k(ranked_list, ground_truth, k=5):
    dcg = 0.0
    for i in range(min(len(ranked_list), k)):
        if ranked_list[i] in ground_truth:
            dcg += 1 / np.log2(i + 2)
    ideal_dcg = sum(1 / np.log2(i + 2) for i in range(min(len(ground_truth), k)))
    return dcg / ideal_dcg if ideal_dcg > 0 else 0
```

```python
hit_scores = []
recall_scores = []
precision_scores = []
ndcg_scores = []

for path in paths:
    user = path[0]
    recommended_songs = [node for node in path if node in df_songs['song_id'].values]

    if user in ground_truth:
        true_songs = ground_truth[user]

        # Compute metrics
        hit = hit_at_k(recommended_songs, true_songs, k=5)
        recall = recall_at_k(recommended_songs, true_songs, k=5)
        precision = precision_at_k(recommended_songs, true_songs, k=5)
        ndcg = ndcg_at_k(recommended_songs, true_songs, k=5)

        # Save scores
        hit_scores.append(hit)
        recall_scores.append(recall)
        precision_scores.append(precision)
        ndcg_scores.append(ndcg)

# Final Averages
avg_hit = sum(hit_scores) / len(hit_scores)
avg_recall = sum(recall_scores) / len(recall_scores)
avg_precision = sum(precision_scores) / len(precision_scores)
avg_ndcg = sum(ndcg_scores) / len(ndcg_scores)

# Print all scores
print(f"🎯 Hit@5:        {avg_hit:.2f}")
print(f"📉 Recall@5:     {avg_recall:.2f}")
print(f"✅ Precision@5:  {avg_precision:.2f}")
print(f"📊 NDCG@5:       {avg_ndcg:.2f}")
```

```python
from collections import defaultdict
import re

segment_scores = defaultdict(lambda: {'hit': [], 'recall': [], 'precision': []})

for path in paths:
    user = path[0]
    recommended = [node for node in path if node in df_songs['song_id'].values]

    if user in ground_truth:
        gt = ground_truth[user]
        segment = re.sub(r'^user_', '', user)  # Extract segment like "20-35_Female"

        segment_scores[segment]['hit'].append(hit_at_k(recommended, gt, k=5))
        segment_scores[segment]['recall'].append(recall_at_k(recommended, gt, k=5))
        segment_scores[segment]['precision'].append(precision_at_k(recommended, gt, k=5))

# Print summary
print("📊 Breakdown by User Segment:")
for seg, metrics in segment_scores.items():
    hit = sum(metrics['hit']) / len(metrics['hit']) if metrics['hit'] else 0
    recall = sum(metrics['recall']) / len(metrics['recall']) if metrics['recall'] else 0
    precision = sum(metrics['precision']) / len(metrics['precision']) if metrics['precision'] else 0
    print(f"{seg:15s} → Hit@5: {hit:.2f} | Recall@5: {recall:.2f} | Precision@5: {precision:.2f}")
```

```python
mood_scores = defaultdict(lambda: {'hit': [], 'recall': [], 'precision': []})
user_moods = df_users.set_index('user_id')['music_Influencial_mood'].to_dict()

for path in paths:
    user = path[0]
    recommended = [node for node in path if node in df_songs['song_id'].values]

    if user in ground_truth and user in user_moods:
        gt = ground_truth[user]
        mood = user_moods[user]

        mood_scores[mood]['hit'].append(hit_at_k(recommended, gt, k=5))
        mood_scores[mood]['recall'].append(recall_at_k(recommended, gt, k=5))
        mood_scores[mood]['precision'].append(precision_at_k(recommended, gt, k=5))

# Print summary
print("\n🎧 Breakdown by User Mood:")
for mood, metrics in mood_scores.items():
    hit = sum(metrics['hit']) / len(metrics['hit']) if metrics['hit'] else 0
    recall = sum(metrics['recall']) / len(metrics['recall']) if metrics['recall'] else 0
    precision = sum(metrics['precision']) / len(metrics['precision']) if metrics['precision'] else 0
    print(f"{mood:35s} → Hit@5: {hit:.2f} | Recall@5: {recall:.2f} | Precision@5: {precision:.2f}")
```

```python
# Count how many users had each song in their ground truth
from collections import Counter

song_counts = Counter([song for songs in ground_truth.values() for song in songs])
popularity = {}

# Classify songs into tiers: high / medium / low
for song, count in song_counts.items():
    if count >= 20:
        popularity[song] = 'High'
    elif count >= 5:
        popularity[song] = 'Medium'
    else:
        popularity[song] = 'Low'

popularity_scores = defaultdict(lambda: {'hit': [], 'recall': [], 'precision': []})

for path in paths:
    user = path[0]
    recommended = [node for node in path if node in df_songs['song_id'].values]

    if user in ground_truth:
        gt = ground_truth[user]
        for song in recommended[:5]:
            tier = popularity.get(song, 'Unknown')
            popularity_scores[tier]['hit'].append(int(song in gt))
            popularity_scores[tier]['recall'].append(recall_at_k([song], gt, k=1))
            popularity_scores[tier]['precision'].append(precision_at_k([song], gt, k=1))

# Print summary
print("\n🔥 Breakdown by Song Popularity Tier:")
for tier, metrics in popularity_scores.items():
    hit = sum(metrics['hit']) / len(metrics['hit']) if metrics['hit'] else 0
    recall = sum(metrics['recall']) / len(metrics['recall']) if metrics['recall'] else 0
    precision = sum(metrics['precision']) / len(metrics['precision']) if metrics['precision'] else 0
    print(f"{tier:10s} → Hit@5: {hit:.2f} | Recall@5: {recall:.2f} | Precision@5: {precision:.2f}")
```
